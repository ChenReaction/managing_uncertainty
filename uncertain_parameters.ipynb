```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams.update({'font.size': 22})
```

```python
T = 1000
E = 415 - 0.023*T
print('E mean: ', E)
print('E three standard deviations: ', 0.03*E)
nu = 0.160 - 2.62e-6*T
print('nu mean: ', nu)
print('nu three standard deviations: ', 0.25*E)
```

```python
E_samples = E + 0.03/3*E*np.random.randn(6000)
nu_samples = nu + 0.25/3*nu*np.random.randn(6000)
```

```python
plt.hist(E_samples,50,density=True)
plt.xlabel(r'E (GPa)')
plt.ylabel(r'probability density')
plt.savefig('E.png',dpi=300,bbox_inches='tight')
```

```python
plt.hist(nu_samples,50,density=True,color='g')
plt.xlabel(r'$\nu}$')
plt.ylabel(r'probability density')
plt.savefig('nu.png',dpi=300,bbox_inches='tight')
```
