```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams.update({'font.size': 14})
```

```python
T = np.linspace(0,1600,9)
print(T)
beta = 87.6
np.exp(5.784)

FS = 359+beta/(1+np.exp(-0.012*(T-1020.67)))  # mean values of function
print('mean values of function: ', FS)
FS_upper_15_percent = 1.15*359+1.15*beta/(1+np.exp(-0.012*(T-1020.67)))
print('upper 15% error bound: ', FS_upper_15_percent)
FS_lower_15_percent = 0.85*359+0.85*beta/(1+np.exp(-0.012*(T-1020.67)))
print('lower 15% error bound: ', FS_lower_15_percent)
FS_shifted_left = 359+beta/(1+np.exp(-0.012*(T-520.67)))
print('upper 15% error bound minus mean: ', FS_upper_15_percent - FS)

centered_synth_exp = FS+20*np.random.randn(9) #standard deviation 20 MPa
print('centered synthetic: ', centered_synth_exp)
shifted_up_exp = FS+25+20*np.random.randn(9)
print('shifted up: ', shifted_up_exp)
shifted_down_exp = FS-25+20*np.random.randn(9)
print('shifted down: ', shifted_down_exp)
shifted_left_synth_exp = FS_shifted_left+20*np.random.randn(9) 
print('shifted left synthetic: ', shifted_left_synth_exp)
```

```python
plt.plot(T,FS,'b',label = 'model mean')
plt.plot(T,FS_upper_15_percent,'b--', label = '15% bound')
plt.plot(T,FS_lower_15_percent,'b--')
plt.legend(loc='upper left', frameon=False)
plt.ylabel(r'flexural strength (MPa)')
plt.xlabel(r'temperature ($\degree$C)')
plt.ylim([300,525])
plt.savefig('flexural_strength_prior_func_only.png',dpi=300,bbox_inches='tight')
```
