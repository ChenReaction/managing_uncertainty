```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2.5,100)

def weib(x,n,a):
    return (a / n) * (x / n)**(a - 1) * np.exp(-(x / n)**a)

loc = 1
scale = 1

plt.plot(x, weib(x, loc, scale))
#plt.plot(x,np.exp(-x)) #equivalent to this when loc=1, scale=1
plt.xlim([0, 2.5])
plt.ylim([0, 1])
plt.xlabel(r'$\zeta$')
plt.ylabel('probability density')
plt.savefig('Weibull_prior_pdf.png',dpi=300)
```
